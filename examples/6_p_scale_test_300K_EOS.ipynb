```python
v0 = 74.698
```

```python
v = v0 * (eta) 
temp = 3000.
```

```python
p = test_EOS.cal_p(v, temp * np.ones_like(v))
p_300 = test_EOS.cal_p(v, 300. * np.ones_like(v))
```

```python
print('for T = ', temp)
for eta_i, p_i, p300_i in zip(eta, p, p_300):
    print("{0: .3f} {1: .2f} {2: .2f}".format(eta_i, p_i, p300_i))
```
